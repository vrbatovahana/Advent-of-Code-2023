```python
df = pd.read_csv(cesta_k_csv, header=None)
df.columns = ['String']
df['String'] = df['String'].apply(lambda x: '.' + x + '...')
novy_radek = pd.DataFrame({'String': ['.' * 144]})
df = pd.concat([novy_radek,df,novy_radek], ignore_index=True)
symboly = {'#', '$', '%', '&', '*', '+', '-', '/', '=', '@'}
part_number = []
for i in range(0, 142):
    for j in range(0, 144):
        # Kontrola délky řetězce před přístupem k jednotlivým znakům
        if j < len(df['String'][i]) and not df['String'][i][j].isdigit():
            continue
        if j + 1 < len(df['String'][i]) and not df['String'][i][j + 1].isdigit():
            # jednociferné
            if (
                (
                    any(symbol in df['String'][i - 1][j] for symbol in symboly)
                    or any(symbol in df['String'][i - 1][j - 1] for symbol in symboly)
                    or any(symbol in df['String'][i - 1][j + 1] for symbol in symboly)
                    or any(symbol in df['String'][i + 1][j] for symbol in symboly)
                    or any(symbol in df['String'][i + 1][j - 1] for symbol in symboly)
                    or any(symbol in df['String'][i + 1][j + 1] for symbol in symboly)
                    or any(symbol in df['String'][i][j - 1] for symbol in symboly)
                    or any(symbol in df['String'][i][j + 1] for symbol in symboly)
                )
                and not j - 1 < 0 and not df['String'][i][j - 1].isdigit()
            ):
                part_number.append(df['String'][i][j])
            continue
        if j + 2 < len(df['String'][i]) and not df['String'][i][j + 2].isdigit():
            # dvouciferné
            if (
                (
                    any(symbol in df['String'][i - 1][j] for symbol in symboly)
                    or any(symbol in df['String'][i - 1][j - 1] for symbol in symboly)
                    or any(symbol in df['String'][i - 1][j + 1] for symbol in symboly)
                    or any(symbol in df['String'][i - 1][j + 2] for symbol in symboly)
                    or any(symbol in df['String'][i + 1][j] for symbol in symboly)
                    or any(symbol in df['String'][i + 1][j - 1] for symbol in symboly)
                    or any(symbol in df['String'][i + 1][j + 1] for symbol in symboly)
                    or any(symbol in df['String'][i + 1][j + 2] for symbol in symboly)
                    or any(symbol in df['String'][i][j - 1] for symbol in symboly)
                    or any(symbol in df['String'][i][j + 2] for symbol in symboly)
                )
                and not j - 2 < 0 and not df['String'][i][j - 2].isdigit()
            ):
                part_number.append(df['String'][i][j : j + 2])
            continue
        # trojciferné
        if (
            (
                any(symbol in df['String'][i - 1][j] for symbol in symboly)
                or any(symbol in df['String'][i - 1][j - 1] for symbol in symboly)
                or any(symbol in df['String'][i - 1][j + 1] for symbol in symboly)
                or any(symbol in df['String'][i - 1][j + 2] for symbol in symboly)
                or any(symbol in df['String'][i - 1][j + 3] for symbol in symboly)
                or any(symbol in df['String'][i + 1][j] for symbol in symboly)
                or any(symbol in df['String'][i + 1][j - 1] for symbol in symboly)
                or any(symbol in df['String'][i + 1][j + 1] for symbol in symboly)
                or any(symbol in df['String'][i + 1][j + 2] for symbol in symboly)
                or any(symbol in df['String'][i + 1][j + 3] for symbol in symboly)
                or any(symbol in df['String'][i][j - 1] for symbol in symboly)
                or any(symbol in df['String'][i][j + 3] for symbol in symboly)
            )
            and not j + 3 >= len(df['String'][i])
        ):
            part_number.append(df['String'][i][j : j + 3])
print(part_number)
```

```python
soucet = sum(int(prvek) for prvek in part_number)
soucet
```
