```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Den6.txt', header=None, sep=':')
df.columns = ['Variable','Value']
df[['Sloupec1', 'Sloupec2', 'Sloupec3', 'Sloupec4']] = df['Value'].str.split(n=3, expand=True)
transposed_df = df.transpose()
transposed_df
```

```python
transposed_df['Win']='|'
transposed_df['Lenght']=np.nan
for i in range(2,6):
    for j in range(1,int(transposed_df.iloc[i, 0])):
        if j*(int(transposed_df.iloc[i, 0]) - j)>int(transposed_df.iloc[i, 1]):
            transposed_df['Win'][i] = transposed_df['Win'][i] + '|'
    transposed_df['Lenght'][i]=len(transposed_df['Win'][i])-1
transposed_df['Lenght'].prod()
```
