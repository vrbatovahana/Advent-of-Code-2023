```python
df=pd.read_csv('Den4.csv', header=None, sep=':')
df.columns=['Card','x']
df[['Winning', 'My']] = df['x'].str.split('|', expand=True)
df['Winning']=df['Winning'].str.strip()
df = df.drop('x', axis=1)
df['My']
```

```python
df['Delka']=''
for i in range(0,197):
    for j in range(0,13):
        if Win[j][i]=='':
            continue
        if Win[j][i] is None:
            continue
# Použijeme regulární výraz k nalezení samostatného výskytu čísla ve sloupci 'My'
        pattern = r'\b' + re.escape(str(Win[j][i])) + r'\b'
        
        if re.search(pattern, str(df['My'][i])):
            df['Delka'][i] = str(df['Delka'][i]) + 'a'
df.tail(20)
```

```python
df['Delka_a']=df['Delka'].apply(lambda x: len(str(x)))
df['Delka_a'].unique()
```

```python
df['Points']=''
for i in range(0,197):
    if df['Delka_a'][i]==0:
        df['Points'][i]=0
    if df['Delka_a'][i]==1:
        df['Points'][i]=2**0
    if df['Delka_a'][i]==2:
        df['Points'][i]=2**1
    if df['Delka_a'][i]==3:
        df['Points'][i]=2**2
    if df['Delka_a'][i]==4:
        df['Points'][i]=2**3
    if df['Delka_a'][i]==5:
        df['Points'][i]=2**4
    if df['Delka_a'][i]==6:
        df['Points'][i]=2**5
    if df['Delka_a'][i]==7:
        df['Points'][i]=2**6
    if df['Delka_a'][i]==8:
        df['Points'][i]=2**7
    if df['Delka_a'][i]==9:
        df['Points'][i]=2**8
    if df['Delka_a'][i]==10:
        df['Points'][i]=2**9
df.tail(20)
```
